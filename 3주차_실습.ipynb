```python
bgr_img = cv2.imread('porygon.jpg') #color, bgr
print(bgr_img.shape) #y:600, x:600, c:3
print(bgr_img.dtype) #uint8 = unsigned char = 1byte
print(bgr_img.size) #=600*600*3 = 1.08MB
```

```python
plt.imshow(cv2.cvtColor(bgr_img,cv2.COLOR_BGR2RGB)) #plt : RGB, cv : BGR
```

```python
#지금 처리한 이미지는 사실 행렬이라 특정 위치로 접근할 수 있다
bgr_img[200,150] #(x,y)=()
```

```python
half_img = bgr_img[::2, ::2] #한 칸 건너 한 칸마다 샘플링한다는 뜻
print(half_img.shape)
plt.imshow(cv2.cvtColor(half_img,cv2.COLOR_BGR2RGB))
plt.show()


# half_img = cv2.resize(bgr_img, (300,300))
# print(half_img.shape)
# plt.imshow(cv2.cvtColor(half_img,cv2.COLOR_BGR2RGB))
```

```python
bgr_img[0:600, 0:600, 0]
#3채널이니까 0,1,2 각각 BGR 순
#blue인 부분만 흰색으로 출력 나머지는 검정
```

```python
plt.imshow(bgr_img[:, :, 2], cmap='gray')
```

```python
#bgr 이미지를 gray 이미지로 바꿔보자

gray_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2GRAY)
print(gray_img.shape)
plt.imshow(gray_img, cmap='gray')
```

```python
# prompt: numpy만을 사용해서 bgr이미지를 gray이미지로 변경해줘

# Calculate grayscale image using NumPy
gray_img_np = 0.2989 * bgr_img[:, :, 2] + 0.5870 * bgr_img[:, :, 1] + 0.1140 * bgr_img[:, :, 0]
gray_img_np = gray_img_np.astype(np.uint8)

# Display the grayscale image using matplotlib
plt.imshow(gray_img_np, cmap='gray')
plt.show()
```

```python
img_zero = np.zeros((600,600), dtype=np.uint8)
img_zero[gray_img < 210] = 255

plt.imshow(img_zero, cmap = 'gray')

#어두운 부분은 0, 밝은 부분은 1(255)로 만들어 라는 함수
#관심있는 부분만 1로 만들고 나머지는 0으로 하는 것
```

```python
#빈 공간을 채우자 > 모폴로지 연산

img_src = img_zero.copy()

img_erode = cv2.erode(img_src, (10,10), iterations=2) #바깥의 노이즈를 제거
img_dilate = cv2.dilate(img_erode, (10,10), iterations=10) #안의 노이즈를 제거

img_dilate
```

```python
#모폴로지 연산의 커널을 직접 지정할 수 있다

# kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))
# print(kernel)


kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3,3))
print(kernel)

img_morph = cv2.morphologyEx(img_src, cv2.MORPH_ERODE, kernel, iterations=1)
img_morph = cv2.morphologyEx(img_morph, cv2.MORPH_CLOSE, kernel, iterations=45) #빈 공간을 다 채우는 MASK 연산, 얼굴 인식에서 사용(CLOSE, OPEN 연산)
plt.imshow(img_morph, cmap='gray')
```

```python
#이미지 내에서 내가 필요한 것들을 뽑아내보자

img_threshold = np.where(img_morph == 255, gray_img, 0).astype(np.uint8)
plt.imshow(img_threshold, cmap = 'gray')



#지금까지 한 것 -> 이미지를 처리해서 RIO(Region of Interest만 남기)
```

```python
# prompt: 위 이미지들(원본이랑 halfsize도 포함해서 img_threshold까지)을 한 번에 보이도록 해줘

import matplotlib.pyplot as plt

# Assuming you have the following variables defined from the previous code:
# bgr_img, half_img, gray_img, gray_img_np, img_zero, img_dilate, img_morph, img_threshold

plt.figure(figsize=(15, 10))  # Adjust figure size for better visualization

# Original Image
plt.subplot(2, 4, 1)
plt.imshow(cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Half-sized Image
plt.subplot(2, 4, 2)
plt.imshow(cv2.cvtColor(half_img, cv2.COLOR_BGR2RGB))
plt.title('Half-sized Image')
plt.axis('off')

# Grayscale Image (cv2.cvtColor)
plt.subplot(2, 4, 3)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image (cv2.cvtColor)')
plt.axis('off')


# Grayscale Image (NumPy calculation)
plt.subplot(2, 4, 4)
plt.imshow(gray_img_np, cmap='gray')
plt.title('Grayscale Image (NumPy)')
plt.axis('off')

# img_zero
plt.subplot(2, 4, 5)
plt.imshow(img_zero, cmap='gray')
plt.title('img_zero')
plt.axis('off')

# img_dilate
plt.subplot(2, 4, 6)
plt.imshow(img_dilate, cmap='gray')
plt.title('img_dilate')
plt.axis('off')

# img_morph
plt.subplot(2, 4, 7)
plt.imshow(img_morph, cmap='gray')
plt.title('img_morph')
plt.axis('off')

# img_threshold
plt.subplot(2, 4, 8)
plt.imshow(img_threshold, cmap='gray')
plt.title('img_threshold')
plt.axis('off')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```
